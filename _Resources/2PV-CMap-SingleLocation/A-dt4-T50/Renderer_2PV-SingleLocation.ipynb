### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Set Mathtext Fontset : Computer Modern
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.clf()
```

### Load Data

```python
mean_error_mat = np.loadtxt("./MeanErrorMuGs.csv", delimiter=",")
diff_error_mat = np.loadtxt("./DiffErrorMuGs.csv", delimiter=",")
```

### Mean Error Color Map

```python
fig_CMap_M = plt.figure(figsize=(8, 8), dpi=200)

ax_CMap_M = fig_CMap_M.add_subplot()
ax_CMap_M.set_title(r'(d) $\bar{E} (\tilde{x}_s (0), \tilde{y}_s (0))$', fontsize=24)

im = ax_CMap_M.imshow(mean_error_mat, aspect=10, cmap='rainbow', extent=[0,250,0,25], vmin=0, vmax=4)

ax_CMap_M.set_xlabel(r'$\Gamma_s$',   fontsize=20)
ax_CMap_M.set_ylabel(r'$\mu$',        fontsize=20)
#ax_CMap_M.set_xlim(0, 250)
#ax_CMap_M.set_ylim(0, 25)

ax_CMap_M.tick_params(labelsize=16)
cb = fig_CMap_M.colorbar(im, ax=ax_CMap_M)

ax_CMap_M.grid(True, 'major', 'both')

ax_CMap_M.legend(fontsize=15, loc="upper right")
fig_CMap_M.savefig("MeanErrorMat.pdf", format="pdf")
fig_CMap_M.savefig("MeanErrorMat.svg", format="svg", transparent=True)
```

### Diff Error Color Map

```python
fig_CMap_D = plt.figure(figsize=(8, 8), dpi=200)

ax_CMap_D = fig_CMap_D.add_subplot()
ax_CMap_D.set_title(r'(d) $\bar{D} (\tilde{x}_s (0), \tilde{y}_s (0))$', fontsize=24)

im = ax_CMap_D.imshow(diff_error_mat, aspect=10, cmap='rainbow', extent=[0,250,0,25], vmin=0, vmax=4)

ax_CMap_D.set_xlabel(r'$\Gamma_s$',   fontsize=20)
ax_CMap_D.set_ylabel(r'$\mu$',        fontsize=20)

ax_CMap_D.tick_params(labelsize=16)
cb = fig_CMap_D.colorbar(im, ax=ax_CMap_D)

ax_CMap_D.grid(True, 'major', 'both')

ax_CMap_D.legend(fontsize=15, loc="upper right")
fig_CMap_D.savefig("DiffErrorMat.pdf", format="pdf")
fig_CMap_D.savefig("DiffErrorMat.svg", format="svg", transparent=True)
```
