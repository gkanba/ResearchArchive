### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Set Mathtext Fontset : Computer Modern
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
```

### Load Data

```python
Traj_ST     = np.loadtxt("./Success-dt3-(0.5-1.5)/Centers_T.csv",     delimiter=",")
Traj_SA     = np.loadtxt("./Success-dt3-(0.5-1.5)/Centers_A.csv",     delimiter=",")
Traj_FT     = np.loadtxt("./Failure-dt3-(0.0-0.0)/Centers_T.csv",     delimiter=",")
Traj_FA     = np.loadtxt("./Failure-dt3-(0.0-0.0)/Centers_A.csv",     delimiter=",")
Error_S     = np.loadtxt("./Success-dt3-(0.5-1.5)/Errors.csv",        delimiter=",")
Error_F     = np.loadtxt("./Failure-dt3-(0.0-0.0)/Errors.csv",        delimiter=",")

T   = 25
dt  = 0.001
CT  = int(np.ceil(T / dt)) + 1
```

### Trajectories Success Small Vortex

```python
t_start     = 0.0
t_end       = 2.75

CT_Start    = int(np.ceil(t_start / dt))
CT_End      = int(np.ceil(t_end / dt))
CT_Width    = CT_End - CT_Start

t_array = np.linspace(t_start, t_end, CT_Width)

fig_all = plt.figure(figsize=(27, 8), dpi=300)

ax_traj_s = fig_all.add_subplot(1, 3, 1)
ax_traj_f = fig_all.add_subplot(1, 3, 2)
ax_traj_e = fig_all.add_subplot(1, 3, 3)

#ax_traj_l.set_title('(a) Trajectory of Large Point Vortex', fontsize=14)
#ax_traj_l.plot(t_array, Traj_T[0, CT_Start:CT_End], "--", c ="#FDA403", label=r'$x_l$', lw=0.4)
#ax_traj_l.plot(t_array, Traj_A[0, CT_Start:CT_End], c ="#FDA403", label=r'$\tilde{x}_l$', lw=0.4)
#ax_traj_l.plot(t_array, Traj_T[2, CT_Start:CT_End], "--", c ="#898121", label=r'$y_l$', lw=0.4)
#ax_traj_l.plot(t_array, Traj_A[2, CT_Start:CT_End], c ="#898121", label=r'$\tilde{y}_l$', lw=0.4)
#ax_traj_l.set_xlim(t_start, t_end)
#ax_traj_l.set_ylim(-2, 2)
#ax_traj_l.set_xlabel(r'$t$', fontsize=14)
#ax_traj_l.set_ylabel(r'$x, y$', fontsize=14)
#ax_traj_l.legend(fontsize=15, loc='upper right')
#ax_traj_l.grid(True, 'major', 'both')

ax_traj_s.set_title(r'(b1) Success $(\tilde{x}_s (t), \tilde{y}_s (t))$', fontsize=36)
ax_traj_s.plot(t_array, Traj_ST[1, CT_Start:CT_End], "--", c ="#FDA403", label=r'$x_s$',    lw=1)
ax_traj_s.plot(t_array, Traj_SA[1, CT_Start:CT_End], c ="#FDA403", label=r'$\tilde{x}_s$',  lw=1)
ax_traj_s.plot(t_array, Traj_ST[3, CT_Start:CT_End], "--", c ="#898121", label=r'$y_s$',    lw=1)
ax_traj_s.plot(t_array, Traj_SA[3, CT_Start:CT_End], c ="#898121", label=r'$\tilde{y}_s$',  lw=1)
ax_traj_s.set_xlim(t_start, t_end)
ax_traj_s.set_ylim(-3, 3)
ax_traj_s.set_xlabel(r'$t$', fontsize=36)
ax_traj_s.set_ylabel(r'$x, y$', fontsize=36)
ax_traj_s.legend(fontsize=28, loc='upper right')
ax_traj_s.grid(True, 'major', 'both')

ax_traj_f.set_title(r'(b2) Failure $(\tilde{x}_s (t), \tilde{y}_s (t))$', fontsize=36)
ax_traj_f.plot(t_array, Traj_FT[1, CT_Start:CT_End], "--", c ="#FDA403", label=r'$x_s$',    lw=1)
ax_traj_f.plot(t_array, Traj_FA[1, CT_Start:CT_End], c ="#FDA403", label=r'$\tilde{x}_s$',  lw=1)
ax_traj_f.plot(t_array, Traj_FT[3, CT_Start:CT_End], "--", c ="#898121", label=r'$y_s$',    lw=1)
ax_traj_f.plot(t_array, Traj_FA[3, CT_Start:CT_End], c ="#898121", label=r'$\tilde{y}_s$',  lw=1)
ax_traj_f.set_xlim(t_start, t_end)
ax_traj_f.set_ylim(-3, 3)
ax_traj_f.set_xlabel(r'$t$', fontsize=36)
ax_traj_f.set_ylabel(r'$x, y$', fontsize=36)
ax_traj_f.legend(fontsize=28, loc='upper right')
ax_traj_f.grid(True, 'major', 'both')


t_array_E = np.linspace(0, 15, 15001)

ax_traj_e.set_title(r'(c) $E(t)$ of both cases', fontsize=36)
ax_traj_e.plot(t_array_E, Error_S[0:15001], label=r'||$E(t)$|| of Success case', lw=1)
ax_traj_e.plot(t_array_E, Error_F[0:15001], label=r'||$E(t)$|| of Failure case', lw=1)
ax_traj_e.plot(t_array_E, 10**(-0.5) * np.ones(shape=(15001, 1)), label=r"$\eta$ : threshold") 

ax_traj_e.set_yscale('log')
ax_traj_e.set_xlim(0, 15)
ax_traj_e.set_xlabel(r'$t$', fontsize=30)
ax_traj_e.set_ylabel(r'$E(t)$', fontsize=30)
ax_traj_e.legend(fontsize=15)
ax_traj_e.grid(True, 'major', 'both')

fig_all.savefig("smalltraj.pdf", format="pdf")
fig_all.savefig("smalltraj.svg", format="svg", transparent=True)
```
