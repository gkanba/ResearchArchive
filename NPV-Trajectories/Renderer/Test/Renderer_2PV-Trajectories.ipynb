### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Set Mathtext Fontset : Computer Modern
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
```

### Load Data

```python
Traj_T      = np.loadtxt("./Centers_T.csv",     delimiter=",")
Traj_A      = np.loadtxt("./Centers_A.csv",     delimiter=",")
Error       = np.loadtxt("./Errors.csv",        delimiter=",")

T   = 25
dt  = 0.001
CT  = int(np.ceil(T / dt)) + 1
```

### Trajectories All

```python
t_start     = 0.0
t_end       = 1.0

CT_Start    = int(np.ceil(t_start / dt))
CT_End      = int(np.ceil(t_end / dt))
CT_Width    = CT_End - CT_Start

t_array = np.linspace(t_start, t_end, CT_Width)

fig_all = plt.figure(figsize=(16, 10), dpi=300)

gs = fig_all.add_gridspec(2, 2, width_ratios=[1.8, 1.5])

ax_traj_l = fig_all.add_subplot(gs[0])
ax_traj_s = fig_all.add_subplot(gs[2])
ax_error  = fig_all.add_subplot(gs[:, 1])

ax_traj_l.set_title('(a) Trajectory of Large Point Vortex', fontsize=14)
ax_traj_l.plot(t_array, Traj_T[0, CT_Start:CT_End], "--", c ="#FDA403", label=r'$x_l$', lw=0.4)
ax_traj_l.plot(t_array, Traj_A[0, CT_Start:CT_End], c ="#FDA403", label=r'$\tilde{x}_l$', lw=0.4)
ax_traj_l.plot(t_array, Traj_T[2, CT_Start:CT_End], "--", c ="#898121", label=r'$y_l$', lw=0.4)
ax_traj_l.plot(t_array, Traj_A[2, CT_Start:CT_End], c ="#898121", label=r'$\tilde{y}_l$', lw=0.4)
ax_traj_l.set_xlim(t_start, t_end)
ax_traj_l.set_ylim(-2, 2)
ax_traj_l.set_xlabel(r'$t$', fontsize=14)
ax_traj_l.set_ylabel(r'$x, y$', fontsize=14)
ax_traj_l.legend(fontsize=15, loc='upper right')
ax_traj_l.grid(True, 'major', 'both')

ax_traj_s.set_title('(b) Trajectory of Small Point Vortex', fontsize=14)
ax_traj_s.plot(t_array, Traj_T[1, CT_Start:CT_End], "--", c ="#FDA403", label=r'$x_s$', lw=0.4)
ax_traj_s.plot(t_array, Traj_A[1, CT_Start:CT_End], c ="#FDA403", label=r'$\tilde{x}_s$', lw=0.4)
ax_traj_s.plot(t_array, Traj_T[3, CT_Start:CT_End], "--", c ="#898121", label=r'$y_s$', lw=0.4)
ax_traj_s.plot(t_array, Traj_A[3, CT_Start:CT_End], c ="#898121", label=r'$\tilde{y}_s$', lw=0.4)
ax_traj_s.set_xlim(t_start, t_end)
ax_traj_s.set_ylim(-3, 3)
ax_traj_s.set_xlabel(r'$t$', fontsize=14)
ax_traj_s.set_ylabel(r'$x, y$', fontsize=14)
ax_traj_s.legend(fontsize=15, loc='upper right')
ax_traj_s.grid(True, 'major', 'both')

ax_error.set_title('(c) Errors of each system', fontsize=14)
ax_error.plot(t_array, Error[CT_Start:CT_End], label=r'||$\mathrm{X} - \hat{\mathrm{X}}$||', lw=0.8)

ax_error.set_yscale('log')
ax_error.set_xlim(t_start, t_end)
ax_error.set_xlabel(r'$t$', fontsize=14)
ax_error.legend(fontsize=15)
ax_error.grid(True, 'major', 'both')

fig_all.savefig("all.pdf")
```

### Trajecotories

```python
t = np.linspace(0, T, CT)

fig_Traj_1 = plt.figure(figsize=(8, 4), dpi=300)
fig_Traj_2 = plt.figure(figsize=(8, 4), dpi=300)

ax_Traj_1 = fig_Traj_1.add_subplot()
ax_Traj_1.set_title('Trajectory of Large Point Vortex', fontsize=20)
ax_Traj_1.plot(t, Traj_T[0, :], "--", c ="#FDA403" ,label=r'$x_{l}$', lw=0.6)
ax_Traj_1.plot(t, Traj_A[0, :], c ="#FDA403" ,label=r'$\tilde{x}_{l}$', lw=0.6)
ax_Traj_1.plot(t, Traj_T[2, :], "--", c ="#898121" ,label=r'$y_{l}$', lw=0.6)
ax_Traj_1.plot(t, Traj_A[2, :], c ="#898121" ,label=r'$\tilde{y}_{l}$', lw=0.6)
ax_Traj_1.set_xlim(0, T)
ax_Traj_1.set_ylim(-4, 4)
ax_Traj_1.set_xlabel(r'$t$', fontsize=18)
ax_Traj_1.set_ylabel(r'$x, y$', fontsize=18)
ax_Traj_1.legend(fontsize=15, loc='upper right')
ax_Traj_1.grid(True, 'major', 'both')

ax_Traj_2 = fig_Traj_2.add_subplot()
ax_Traj_2.set_title('Trajectory of Small Point Vortex', fontsize=20)
ax_Traj_2.plot(t, Traj_T[1, :], "--", c ="#FDA403" ,label=r'$x_s$', lw=0.6)
ax_Traj_2.plot(t, Traj_A[1, :], c ="#FDA403" ,label=r'$\tilde{x}_s$', lw=0.6)
ax_Traj_2.plot(t, Traj_T[3, :], "--", c ="#898121" ,label=r'$y_s$', lw=0.6)
ax_Traj_2.plot(t, Traj_A[3, :], c ="#898121" ,label=r'$\tilde{y}_s$', lw=0.6)
ax_Traj_2.set_xlim(0, T)
ax_Traj_2.set_ylim(-4, 4)
ax_Traj_2.set_xlabel(r'$t$', fontsize=18)
ax_Traj_2.set_ylabel(r'$x, y$', fontsize=18)
ax_Traj_2.legend(fontsize=15, loc='upper right')
ax_Traj_2.grid(True, 'major', 'both')

fig_Traj_1.savefig('Traj_1.pdf', format='pdf')
fig_Traj_2.savefig('Traj_2.pdf', format='pdf')
```
